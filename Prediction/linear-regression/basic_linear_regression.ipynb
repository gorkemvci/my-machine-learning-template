# Basic Linear Regression
---
**Y = aX + b**

---
LINEAR REGRESSION tries to model the data it uses for training on the y = ax + b  line.

```
import pandas as pd
data = pd.read_csv("data.csv")
#print(data)
months= data[["Months"]]
sales= data[["Sales"]]
#print(months)
#print(sales)
```

# Splitting the dataset for training
We divide our data into x_train, x_test , y_train , y_test as test_size 0.33.


```
from sklearn.model_selection import train_test_split
x_train, x_test , y_train , y_test = train_test_split(months,sales,test_size=0.33, random_state=0)
# print(x_train)
# print(y_train)
# print(x_test)
# print(y_test)
```

# LinearRegression Model
We create our model thanks to the linear regression model.
*   predict= These values are the estimates of our model.
*   y_test= These values are the actual values.




```
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(x_train,y_train)
predict =lr.predict(x_test)
#print(predict)
#print(y_test)
```

# We visualized it with the help of matplotlib 
The values indicated by the green line are the actual values of our data. The values shown in blue are the values predicted by our model.

```
import matplotlib.pyplot as plt
x_train = x_train.sort_index()
y_train = y_train.sort_index()
plt.plot(x_train,y_train,color="green")
plt.plot(x_test ,predict, color="blue")
```
