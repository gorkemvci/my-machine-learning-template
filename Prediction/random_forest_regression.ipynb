# Random Forest Regression
Random Forest Regression  is a variation of the decision tree algorithm, and it adds randomness when constructing trees. It divides the data to classify each partition homogeneously. However, the selection of features at each node is made randomly from a subset of features, creating different branches for the tree and reducing the problem of overfitting.

# Intro
We add the libraries to be used.
We are reading the data from csv file with the help of pandas.


```python
import numpy
import matplotlib.pyplot as plt
import pandas
data= pandas.read_csv("salary_data.csv")
#print(data)
x =data.iloc[:,-2:-1]
y =data.iloc[:,-1:]
#print(x)
#print(y)
```

# Linear Regression
It is not suitable for this data set as it is modeled with a straight line in LINEAR REGRESSION.

```python
from sklearn.linear_model import LinearRegression
lin_reg = LinearRegression()
lin_reg.fit(x,y)
X=x.values
Y=y.values
plt.scatter(X,Y , color ="red")
plt.plot(x, lin_reg.predict(X), color = "blue")
```

# Random Forest Regression
By giving the n_estimators value 10, we have created 10 different decision trees. Thus preventing memorization. You can get different values by reducing or increasing the value of n_estimators.

```python
from sklearn.ensemble import RandomForestRegressor
rf_reg= RandomForestRegressor(n_estimators=10, random_state=0)
rf_reg = rf_reg.fit(x,y)
plt.scatter(x,y, color ="red")
plt.plot(x, rf_reg.predict(x), color = "blue")
```
