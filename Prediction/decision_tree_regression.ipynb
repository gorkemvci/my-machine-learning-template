# Decision Tree Regression
It represents data in the form of a tree structure, and each node is associated with a property or attribute. The values of these properties determine the branches and leaves in the decision tree. The algorithm divides the data so that each partition is classified homogeneously.

# Intro
We add the libraries to be used.
We are reading the data from csv file with the help of pandas.


```python
import numpy
import matplotlib.pyplot as plt
import pandas
data= pandas.read_csv("salary_data.csv")
#print(data)
x =data.iloc[:,-2:-1]
y =data.iloc[:,-1:]
#print(x)
#print(y)
```

# Linear Regression
It is not suitable for this data set as it is modeled with a straight line in LINEAR REGRESSION.

```python
from sklearn.linear_model import LinearRegression
lin_reg = LinearRegression()
lin_reg.fit(x,y)
X=x.values
Y=y.values
plt.scatter(X,Y , color ="red")
plt.plot(x, lin_reg.predict(X), color = "blue")
```

# Decision Tree Regression


```python
from sklearn.tree import DecisionTreeRegressor
dt_r= DecisionTreeRegressor(random_state=0)
dt_r = dt_r.fit(x,y)
plt.scatter(x,y, color ="red")
plt.plot(x, dt_r.predict(x), color = "blue")
```
